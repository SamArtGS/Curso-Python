Python tiene un paquete incorporado llamado **re**, que se puede usar para trabajar con expresiones regulares.
Puedes comenzar a usar expresiones regulares usando el m칩dulo **re**.

**Funciones incluidas en el paquete:**


- `findall`: Devuelve una lista que contiene todas las coincidencias
- `search`:	 Devuelve un objeto Match si hay una coincidencia en cualquier parte de la cadena
- `split`:	 Devuelve una lista donde la cadena se ha dividido en cada coincidencia
- `sub`:	 Reemplaza una o varias coincidencias con una cadena
- `match`:	 Determina si la expresi칩n coincide al principio de la cadena

```python
import re

p = re.compile(r'\d+')

x = p.findall('11 es un n칰mero y 12 tambi칠n, despu칠s tenemos al 13, 14, 15 y sin olvidar al 20')

print(x)

if x:
    print("Hay al menos una coincidencia!")
else:
    print("No hay coincidencias")
```

## Contestemos las preguntas iniciales con un archivo externo

### 1. 쮺u치ntas conexiones se realizaron del 12 al 15 de Diciembre por /administrator y con el m칠todo POST?

**Expresi칩n regular:** `\d+\.\d+\.\d+\.\d+ - - \[1[2-5]\/Dec\/\d+:.{17}POST \/administrator\/.+`

```python
import re

#Compilamos nuestra expresi칩n regular
miExpresionRegular = re.compile(r'\d+\.\d+\.\d+\.\d+ - - \[1[2-5]\/Dec\/\d+:.{17}POST \/administrator\/.+')

#Al abrir el archivo apache.log como 'archivo' has esto:

with open('apache.log') as archivo:
    contenido = archivo.read()
    print("N칰mero de conexiones del 12-15 Dic POST por arministrador: ",len(miExpresionRegular.findall(contenido)))
    print("\nEjemplos: \n")
    print(miExpresionRegular.findall(contenido)[0]+"\n")
    print(miExpresionRegular.findall(contenido)[1])
```

### 2. 쮺u치ntas conexiones GET de usuarios que no sean /administrator se realizaron en horario de 12 am a 9 am?

**Expresi칩n regular:** `\d+\.\d+\.\d+\.\d+ - - \[\d+\/\w+\/\d+:0[0-9]:\d+:\d+ \+\d+\] "GET \/(?!administrator).+`

```python
import re

#Compilamos nuestra expresi칩n regular
miExpresionRegular2 = re.compile(r'\d+\.\d+\.\d+\.\d+ - - \[\d+\/\w+\/\d+:0[0-9]:\d+:\d+ \+\d+\] "GET \/(?!administrator).+')

#Al abrir el archivo apache.log como 'archivo' has esto:

with open('apache.log') as archivo:
    contenido = archivo.read() # Lee su contenido
    print("N칰mero de conexiones GET no realizadas por /administrator de 0 am a 9 am: ",len(miExpresionRegular2.findall(contenido))) #Imprime el tama침o del arreglo de todas las coincidencias
    print("\nEjemplos: \n") 
    print(miExpresionRegular2.findall(contenido)[0]+"\n")
    print(miExpresionRegular2.findall(contenido)[1])
```

### 3. 쮺u치ntas conexiones fueron exitosas (C칩digo 200) el 13 y 14 de Diciembre?

**Expresi칩n regular:** `\d+\.\d+\.\d+\.\d+ - - \[1[3-4]\/Dec\/\d+:\d+:\d+:\d+ \+\d+\] "\w+ [^ ]+ [^ ]+ 200 .+`

```python
import re

#Compilamos nuestra expresi칩n regular
miExpresionRegular3 = re.compile(r'\d+\.\d+\.\d+\.\d+ - - \[1[3-4]\/Dec\/\d+:\d+:\d+:\d+ \+\d+\] "\w+ [^ ]+ [^ ]+ 200 .+')

#Al abrir el archivo apache.log como 'archivo' has esto:

with open('apache.log') as archivo:
    contenido = archivo.read() # Lee su contenido
    print("Conexiones Exitosas (200) el 13 y 14 de Diciembre: ",len(miExpresionRegular3.findall(contenido))) #Imprime el tama침o del arreglo de todas las coincidencias
    print("\nEjemplos: \n") 
    print(miExpresionRegular3.findall(contenido)[0]+"\n")
    print(miExpresionRegular3.findall(contenido)[1])
```
