## Ejemplo de m칠todos num칠ricos 游꼱

**<span style="color:red"> ** IMPORTANTE ** </span>**

Para escribir una ecuaci칩n tienes que separar todo y multiplicar con \*:  
f=2x+3 仇     
f = 2*x + 3 九

Para poner una potencia por ejemplo $2^3$ es poniendo doble aster칤sco. Ejemplo:

$$f(x) = x^2 + 2x +1$$

Equivale a:

````python
import sympy as sy
from sympy.functions import sin,cos
import math
import matplotlib.pyplot as plt
x = sy.symbols('x')
def taylor(funcion,c,n):
    i = 0
    p = 0
    while i <= n:
        p = p + (funcion.diff(x,i).subs(x,c))/(math.factorial(i))*(x-c)**i
        i += 1
    return p
````

````python
#CAMBIA LA FUNCION, en d칩nde est치 alineado taylor (este caso c=0) y cu치l es el mayor exponenete que quieres (este es 4)
p = taylor(cos(x),0,4)
print("El polinomios de Taylor es: ",p)
````

## M칠todo de bisecci칩n 游늵

````python
# Inserta la ecuaci칩n a buscar ra칤z
fx = lambda x: x**2 + 9*x + 4
````

````python
# Inserta tu rango y tolerancia
a = -5
b = 2
tolera = 0.01
````

````python
# PROCEDIMIENTO
tabla = []
tramo = b-a
fa = fx(a)
fb = fx(b)
i = 1
while (tramo>tolera):
    c = (a+b)/2
    fc = fx(c)
    tabla.append([i,a,c,b,fa,fc,fb,tramo])
    i = i+1
                 
    cambia = np.sign(fa)*np.sign(fc)
    if (cambia<0):
        b = c
        fb = fc
    else:
        a=c
        fa = fc
    tramo = b-a
c = (a+b)/2
fc = fx(c)
tabla.append([i,a,c,b,fa,fc,fb,tramo])
tabla = np.array(tabla)
raiz = c
np.set_printoptions(precision = 4)
print('i   a     c    b     f(a)   f(c)  f(b)  toler')
n=len(tabla)
for i in range(0,n,1):
    unafila = tabla[i]
    formato = '{:.0f}'+' '+(len(unafila)-1)*'{:.3f} '
    unafila = formato.format(*unafila)
    print(unafila)
print('raiz: ',raiz)
````

### M칠todo de Newton Raphson 救넖잺

````python
x=sy.symbols('x')
#funcion
funcion=sin(x) + cos(x)

derivada=sy.diff(funcion,x)

x_0=0
xr=x_0

ea=100/100

es=0.001/100
contador=-1
print("i\t   xi\t     Error Absoluto Porcentual")
while ea>es:
    xra=xr
    contador+=1
    newton_rhapson=x-(funcion/derivada)

    xr=newton_rhapson.evalf(subs={x: xr})
    
    #error aproximado relativo porcentual
    ea=sy.Abs(((xr-xra)/xr)*100)
    #resultado
    print(contador   ,xra, ea)
````

## M칠todo de la regla falsa.

## Regresi칩n 游꼢

La expresi칩n matem치tica general de una regresi칩n lineal es:


$$y = \beta_0 + \beta_1x_1 + \beta_2x_2 + ... + \beta_nx_n + \epsilon$$

## Regresi칩n: Regresi칩n Lineal Simple

Cuando solo tenemos la expresi칩n con i = 1 tendremos la expresi칩n dada en:

$$\beta_0 + \beta_1x_1 + \epsilon$$

Los que tratamos de hacer es: Obtener las betas y epsilon. Para quedar m치s claro imagen la expresi칩n anterior como:

$$y = mx + b$$

Si recordar치n de su curso de matem치ticas en la secundaria m es la pendiente y b la ordenada al origen y podemos decir que beta 0 es equivalente a b y beta 1 es equivalente a beta 1.

Ahora.쮺칩mo calculaban la pendiente?

$$ m = \frac{y_2-y_1}{x_2-x_1}$$

#### M칤nimos cuadrados
Para poder determinar la recta que tenga mejor ajuste necesitaremos una t칠cnica llamada m칤nimos cuadrados y esto como para qu칠?

Bueno lo que hace es minimizar el error cuadr치tico medio dado:

$$b = \frac{\sum y\sum x^2 - \sum x-\sum x y}{N\sum x^2 - (\sum x)^2}$$

$$ m = \frac{N\sum xy - \sum x\sum y}{N\sum x^2 - (\sum x)^2} $$

**Peeeero**, imag칤nense con un dataset enorme, gigantesco donde la memoria es primordial. 쯌olver칤an a ejecutar estas ecuaciones para cada nuevo dato que tengamos en el dataset?

**Pues la respuesta es NO**. Para ello existen los m칠todos num칠ricos.

Para calcular b y m no es necesario seguir esas ecuaciones, ya que son muy precisas. 

> Los que ya hayan llevado m칠todos num칠ricos me entender치n.

Cuando una computadora realiza c치lculos siempre lo hace con m칠todos num칠ricos y estos tienen una tolerancia y mediante iteraciones vamos reduciendo esa tolerancia lo m치s posible siendo un tiempo determinado. Se aceptan errores, claro, y esto nos permite que si ya tenemos ajustado nuestro modelo solo tengamos que moverlo un poquito.




## Tama침os de predicci칩n. 

Por lo general en el 치rea de estad칤stica existe algo llamado estad칤stica inferencial donde tomamos de una muestra de la poblaci칩n. Considerarlo al momento de obtener muestras muy peque침as.

Para calcular ese error o tolerancia se recurre a algo llamado el c치lculo del error cuadr치tico medio de estad칤stica que quiere decir, la suma de todas las diferencias de la funci칩n calculada hasta los inputs cu치nto me da. Buscamos que se la m칤nima posible y al ser una funci칩n cuadr치tica esta es una par치bola.


$$ECM(\theta') = E[(\theta' - \theta)^2] = Var(\theta') + sesgo(\theta') ^2$$

````python
import numpy as np
import random
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

%matplotlib inline
 
# Generador de distribuci칩n de datos para regresi칩n lineal simple
def generador_datos_simple(beta, muestras, desviacion):
    # Genero n (muestras) valores de x aleatorios entre 0 y 100
    x = np.random.random(muestras) * 100
    # Genero un error aleatorio gaussiano con desviaci칩n t칤pica (desviacion)
    e = np.random.randn(muestras) * desviacion
    # Obtengo el y real como x*beta + error
    y = x * beta + e
    return x.reshape((muestras,1)), y.reshape((muestras,1))
 
# Par치metros de la distribuci칩n normal
desviacion = 200
beta = 10
n = 50

x, y = generador_datos_simple(beta, n, desviacion)

plt.scatter(x, y)
plt.title("Gr치fica de puntos aleatorios")
plt.show()
````

````python
# Creo un modelo de regresi칩n lineal
modelo = linear_model.LinearRegression()

# Entreno el modelo con los datos (X,Y)
modelo.fit(x, y)


# Ahora puedo obtener el coeficiente b_1
print ('Coeficiente beta1: ', modelo.coef_[0])
 
# Podemos predecir usando el modelo
y_pred = modelo.predict(x)


# Por 칰ltimo, calculamos el error cuadr치tico medio y el estad칤stico R^2
print ("Error cuadr치tico medio: ", mean_squared_error(y, y_pred))

plt.scatter(x, y) ## Graficamos los puntos

plt.plot(x, y_pred, color='red') ## Graficamos la predicci칩n de SciKitLearn

x_real = np.array([0, 100])
y_real = x_real*beta

plt.plot(x_real, y_real, color='green') ## Graficamos con M칠todo de M칤nimos Cuadrados
plt.show()
````

````python
import numpy as np #Importamos numpy para los arreglos
import datetime
import tensorflow as tf # Importamos TensorFlow
from tensorflow.keras.datasets import fashion_mnist ## Importams un dataset de Keras, el MNIST

#Cargar el dataset Fashion Mnist y las variables Xtrain, Xtest, Ytrain Ytest guardan los valores de retorno
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()
````

### Normalizaci칩n de las im치genes

Se divide cada imagen en los conjunto de entrenamiento (train) y de pruebas (test) entre el valor m치ximo de cada uno de los p칤xeles (255), ya que 0 significa negro y 255 significa un pixel blanco.

De este modo, cada p칤xel se hallar치 en el rango [0, 1], siendo 0 negro y 1 blanco. 

- 쯇ara qu칠 hacemos esto? 

Al normalizar las im치genes, nos aseguramos que nuestro modelo de RNA entrenar치 m치s r치pidamente.

````python
 X_train = X_train / 255.0
X_test = X_test / 255.0
````

### Redimensionar el dataset

Como vamos a utilizar una red neuronal totalmente conectada, vamos a redimensionar los subconjuntos de entrenamiento y testing a formato de vector en lugar de en formato de matriz. Esto significa, nuestras im치genes tienen 28x28 pixeles, podemos representar cada uno de sus pixeles como un color entre 0 y 1 ya que normalizamos las im치genes. Ahora en lugar de ponerlo como matriz pondremos un arreglo lineal de 1x(28x28) = 1x784. Esto para tener la primera capa 784 neuronas.

````python
#Como cada imagen tiene 28x28 p칤xeles, usamos la funci칩n reshape en todo el dataset de entrenamiento para convertirlo 
# en vectores de tama침o [-1 (todos los elementos), anchura * altura]
X_train = X_train.reshape(-1, 28*28)
print(X_train.shape)
X_test = X_test.reshape(-1, 28*28)
print(X_test.shape)
````

## Que empiece la construcci칩n del RNA 游

## - Definir el modelo
Simplemente se define un objeto de modelo Sequential.

````python
model = tf.keras.models.Sequential()
````

## - A침adir la primera capa totalmente conectada (capa Densa)

Hyper-parametros de la capa:

- N칰mero de unidades/neuronas: **128**

- Funci칩n de activaci칩n: **ReLU**

````python
model.add(tf.keras.layers.Dense(units=128, activation='relu', input_shape=(784, )))
````

## - A침adir una capa de Dropout

Dropout es una t칠cnica de Regularization donde aleatoriamente se asignan a ciertas neuronas de la red el valor cero. De este modo, mientras se entrena, estas neuronas no actualizar치n sus valores. Al tener cierto porcentaje de neuronas sin actualizar, el proceso de entrenamiento toma m치s tiempo pero por contra tenemos menos posibilidades de sufrir sobreentrenamiento (overfitting).

````python
model.add(tf.keras.layers.Dropout(0.2))
````

## - A침adir capa de salida

En el caso de MNIST tenemos 10 posibles prendas de ropa, por lo que a침adiremos una capa con 10 neuronas 칰nicamente al final, en este caso como es una capa de salida y queremos probabilidades por lo mismo, utilizaremos una funci칩n de activaci칩n SoftMax

- N칰mero de unidades/neuronas: **10** en el caso del Fashion MNIST

- Funci칩n de activaci칩n: **softmax** (probabilidades de cada clase)

````python
model.add(tf.keras.layers.Dense(units=10, activation='softmax'))
````

## Compilar el modelo

Ahora compilaremos todo nuestro modelo, para ello es necesario una funci칩n de 

- Funci칩n optimizaci칩n: Adam

- Funci칩n de costo o p칠rdida: Sparse softmax (categorical) crossentropy

- M칠trica: sparse_categorical_accuracy comprueba si el valor verdadero m치ximo es igual al 칤ndice del valor predicho m치ximo.

````python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['sparse_categorical_accuracy'])
model.summary()
````

## Entrenar el modelo


````python
model.fit(X_train, y_train, epochs=5)
````

# Evaluaci칩n del modelo

````python
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print("Precisi칩n del modelo: ",test_accuracy)
````
